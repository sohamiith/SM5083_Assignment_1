```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.linspace( -0.7 , 0.7 , 200 )
y = np.linspace( -0.7 , 0.7 , 200 )
 
a, b = np.meshgrid( x , y )
 
C = a ** 2 + b ** 2 - 0.2
 
figure, axes = plt.subplots()
 
axes.contour( a , b , C , [0] )
axes.set_aspect( 1 )

x_values = [-0.4, -0.2]
y_values = [-0.2,0.4]
plt.plot(x_values, y_values, '-r')

x_values = [-0.4, 0.4]
y_values = [-0.2,-0.2]
plt.plot(x_values, y_values, '-b')

x_values = [-0.2, 0.4]
y_values = [0.4,-0.2]
plt.plot(x_values, y_values, '-g')

plt.plot([-0.6,0.6], [-0.6,0.6])
plt.plot([-0.6,0.6], [0.2,-0.2])
plt.plot([0.0,0.0], [-0.6,0.6])

plt.title( 'Circum center of the triangle' )
plt.plot(-0.4, -0.2, 'ro')
plt.plot(-0.2, 0.4, 'ro')
plt.plot(0.4, -0.2, 'ro')
plt.plot(0.0, 0.0, 'ro')
plt.grid()
plt.show()
plt.savefig('assignment_1.png')
```
